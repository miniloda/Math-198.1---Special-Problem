```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dengue_2016 = pd.read_csv('data/dengue/hdx/doh-epi-dengue-cases-2016.csv', skiprows=[1,2])
dengue_2017 = pd.read_csv('data/dengue/hdx/doh-epi-dengue-cases-2017.csv', skiprows=[1,2])
dengue_2018 = pd.read_csv('data/dengue/hdx/doh-epi-dengue-cases-2018.csv', skiprows=[1,2])
dengue_2019 = pd.read_csv('data/dengue/hdx/doh-epi-dengue-cases-2019.csv', skiprows=[1,2])
dengue_2020 = pd.read_csv('data/dengue/hdx/doh-epi-dengue-cases-2020.csv', skiprows=[1,2])
dengue_2021 = pd.read_csv('data/dengue/hdx/doh-epi-dengue-cases-2021.csv', skiprows=[1,2])
```

```python
dengue_2017.head()
```

```python

df = pd.concat([dengue_2016, dengue_2017, dengue_2018, dengue_2019, dengue_2020, dengue_2021], axis = 0) 
```

```python
df[df['Region'].str.upper().str.contains('ILOILO')]
```

```python
df.plot(x='date', y='cases', figsize=(15, 5))
```

```python
df_all_ilo = df[df['loc'].str.upper().str.contains('ILOILO CITY')]
df_all_ilo
```

```python
df_all_ilo.plot(x='date', y='cases', figsize=(15, 5))
plt.title('Dengue Cases in Iloilo City')
plt.xlabel('Date')
plt.ylabel('Count of cases')
```

```python
df_all_ilo.to_csv('data/dengue/hdx/doh-epi-dengue-cases-iloilo.csv', index=False)
```
